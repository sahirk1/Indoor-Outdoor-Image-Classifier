```python
import numpy as np
import cv2
import math
```

Generating indoor samples from the youtube videos (saved to local disk). listed below. The videos are tours of vibrant and diverse houses that include a variety of different subsettings such as bedroom, bathrooms, living rooms, home offices,laundry rooms, etc.

1. https://www.youtube.com/watch?v=jpcnE28qfjQ
2. https://www.youtube.com/watch?v=qH51FvFcKv4
3. https://www.youtube.com/watch?v=PuAN2bPe9xo

Frames were extracted at two second intervals. Every fifth frame was used for the indoor portion of the test set. The remaining were saved to the training set.

```python
indoor1 = cv2.VideoCapture('Documents/Indoor Videos/indoor1')
indoor2 = cv2.VideoCapture('Documents/Indoor Videos/indoor2')
indoor3 = cv2.VideoCapture('Documents/Indoor Videos/indoor3')

framerate = indoor1.get(5)
framecount = 0
count = 0
trainCount = 1
testCount = 1

while(True):
    
    success, image = indoor1.read()

    if not success: break
        
    if framecount % (2 * framerate) == 0:
        
        name = '' 
        
        if count % 5 == 0:
            name = 'Documents/Image Data/Test Data/Indoor/' + str(testCount) + '.jpg'
            testCount += 1
        else:
            name = 'Documents/Image Data/Training Data/Indoor/' + str(trainCount) + '.jpg'
            trainCount += 1
            
        count += 1
        cv2.imwrite(name, image)

    framecount += 1

indoor1.release()

framerate = indoor2.get(5)
framecount = 0

while(True):
 
    success, image = indoor2.read()

    if not success: break
        
    if framecount % (2 * framerate) == 0:
        
        name = '' 
        
        if count % 5 == 0:
            name = 'Documents/Image Data/Test Data/Indoor/' + str(testCount) + '.jpg'
            testCount += 1
        else:
            name = 'Documents/Image Data/Training Data/Indoor/' + str(trainCount) + '.jpg'
            trainCount += 1
            
        count += 1
        cv2.imwrite(name, image)

    framecount += 1

indoor2.release()

framerate = indoor3.get(5)
framecount = 0

while(True):
  
    success, image = indoor3.read()

    if not success: break
        
    if framecount % (2 * framerate) == 0:
        
        name = '' 
        
        if count % 5 == 0:
            name = 'Documents/Image Data/Test Data/Indoor/' + str(testCount) + '.jpg'
            testCount += 1
        else:
            name = 'Documents/Image Data/Training Data/Indoor/' + str(trainCount) + '.jpg'
            trainCount += 1
            
        count += 1
        cv2.imwrite(name, image)

    framecount += 1

indoor3.release()
cv2.destroyAllWindows()
```
